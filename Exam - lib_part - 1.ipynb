```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display
```

```python
df = pd.read_csv("Items.csv")
df.head()
```

```python
df_Items = df.copy()
display(df_Items.info())
display(df_Items.describe().T)
```

```python
df = pd.read_csv("customers.csv")
df.head()
```

```python
df_customers = df.copy()
display(df_customers.info())
display(df_customers.describe().T)
```

```python
df = pd.read_csv("order_item.csv")
df.head()
```

```python
df_order_item = df.copy()
display(df_order_item.info())
display(df_order_item.describe().T)
```

```python
df = pd.read_csv("orders.csv")
df.head()
```

```python
df_orders = df.copy()
display(df_orders.info())
display(df_orders.describe().T)
```

```python
# Question 2 - 

# Information analysis:

# 1.

avg_price = df_Items["item_price"].mean()

display(f"Average item price: {avg_price:.2f}")
```

```python
# 2.

merged = pd.merge(df_order_item, df_orders[['order_id', 'customer_id']], on='order_id', how='inner')

top_customer = merged.groupby('customer_id')['quantity'].sum().reset_index()

top_customer = top_customer.sort_values(by='quantity', ascending=False).head(1)

result = pd.merge(top_customer, df_customers, left_on='customer_id', right_on='id')

display(result[['customer_id', 'first_name', 'last_name', 'quantity']])
```

```python
# Question 2 -

# graph - 

# 1.

cleaned_customers = df_customers[df_customers['gender'].notna()]

gender_counts = cleaned_customers['gender'].value_counts().reset_index()
gender_counts.columns = ['gender', 'count']

plt.figure(figsize=(10, 4))
sns.barplot(data=gender_counts, x='gender', y='count')
plt.title('Number of Customers by Gender')
plt.xlabel('Gender')
plt.ylabel('Number of Customers')
plt.tight_layout()
plt.show()
```

```python
# 2.

cleaned_customers = df_customers[df_customers['nationallity'].notna()]

nationality_counts = cleaned_customers['nationallity'].value_counts().reset_index()
nationality_counts.columns = ['nationallity', 'count']

plt.figure(figsize=(12, 5))
sns.barplot(data=nationality_counts, x='nationallity', y='count')
plt.title('Number of Customers by Nationality')
plt.xlabel('Nationality')
plt.ylabel('Number of Customers')
plt.tight_layout()
plt.show()
```
